```python
import numpy as np
import pandas as pd
import altair as alt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from vega_datasets import data

dfID = pd.read_csv(data.population_engineers_hurricanes.url)

names = list(dfID['state'])[:-1]
stateID = list(dfID['id'])[:-1]
# dfID.head()
```

```python
df = pd.read_csv("election2020.csv")
df = df.sort_values(by=['state']) 
df.insert(0, 'id', stateID)
df["percent_republican"] = 100-df["percent_democrat"]
df.head(10)
```

```python
df['won'] = np.where(df['percent_democrat'] > 50, "Dem", "Rep")
df.head()
```

```python
states = alt.topo_feature(data.us_10m.url, 'states')
domain = ['Dem', "Rep"]
colors = ["steelblue", "indianred"]


alt.Chart(states).mark_geoshape(stroke='white', width=0.01).encode(
    color = alt.Color('won:N', scale=alt.Scale(domain=domain, range=colors)),
    tooltip = ['state:N', "DEM:Q", "REP:Q", 'won:N']
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(df, 'id', ['state', "DEM", "REP",'won'])
).project(
    type='albersUsa'
).properties(
    title="Election Result Map",
    width=800,
    height=480
)
```

```python
states = alt.topo_feature(data.us_10m.url, 'states')

alt.Chart(states).mark_geoshape(stroke='white', width=0.01).encode(
    color = alt.Color('percent_democrat:Q', scale=alt.Scale(scheme="blues")),
    tooltip = ['state:N', alt.Tooltip('percent_democrat:N' ,format='.4')]
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(df, 'id', ['state', 'percent_democrat'])
).project(
    type='albersUsa'
).properties(
    title="Percentage Democrat Map",
    width=800,
    height=480
)
```

```python
df_vac = pd.read_csv("cdc_vaccines_distributed_administered_by_jurisdiction.csv")


df_vac.loc[(df_vac.state_name == 'New York State'),'state_name']="New York"
df_vac = df_vac[df_vac['state_name'].isin(names)]
df_vac = df_vac.sort_values(by=['state_name']) 
df_vac.insert(0, 'id', stateID)
df_vac = df_vac.rename(columns={'doses_administered_moderna':'moderna', 'doses_administered_pfizer':"pfizer",'doses_administered_unknown_manufacturer':"unknown"})

percent_dem = list(df["percent_democrat"])
names2 = list(df["state"])
df_vac["percent_democrat"] = percent_dem


df_vac.head()
```

```python
alt.Chart(states).mark_geoshape(stroke='white', width=0.01).encode(
    color = alt.Color('pct_population_received_at_least_one_dose:Q', scale=alt.Scale(scheme="blues")),
    tooltip = ['state_name:N', alt.Tooltip('pct_population_received_at_least_one_dose:Q', title='% population get >= 1 dose', format='.4')]
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(df_vac, 'id', ['state_name', 'pct_population_received_at_least_one_dose'])
).project(
    type='albersUsa'
).properties(
    title="Percentage Population Received >= 1 Dose",
    width=800,
    height=480
)
```

```python
dropdown = alt.binding_select(options=df_vac['state_name'].unique(), name="Select a state: ")

selection = alt.selection(type="single", fields=['state_name'], bind=dropdown)

alt.Chart(df_vac).mark_circle().encode(
    y = "percent_democrat",
    x = "pct_population_received_at_least_one_dose",
    size = alt.Size("population:Q",scale=alt.Scale(range=[0,1500])),
    color = alt.Color("doses_distributed_per_100k:Q", scale=alt.Scale(scheme="redblue")),
    tooltip = ['state_name', alt.Tooltip('pct_population_received_at_least_one_dose', title='% population get >= 1 dose')],
    opacity = alt.condition(selection, alt.value(1), alt.value(.2))
).add_selection(selection).properties(width=600, height=360).interactive()
```

```python
selection = alt.selection(type="multi", fields=["state_name"])
base = alt.Chart(df_vac).properties(width=300, height=700)

overview = alt.Chart(df_vac).mark_bar().encode(
    y = "state_name:N",
    x = "doses_administered_per_100k:Q",
    color = alt.condition(selection, alt.value("olive"), alt.value("lightgrey"))
).add_selection(selection).properties(title="Adminstered Vaccine per 100K",width=300, height=700)

detail = alt.Chart(df_vac).transform_fold(
  ['moderna', 'pfizer','unknown'],
  as_=['column', 'value']
).mark_bar().encode(
  y='state_name:N',
  x=alt.X('value:Q', title="# adminstered doses"),
  color='column:N'
).transform_filter(selection).properties(title="Vaccine Provider Breakdown",width=300, height=700)


overview | detail
```
